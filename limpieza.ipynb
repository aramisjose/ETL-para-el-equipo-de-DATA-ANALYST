                                                        TRANSFORMACIONES

                                    Transformaciones para el dataframe datos_disney

Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)

```python
## Esta función concatenara la inicial de disney a la columna show_id
def funcion(var):
    return 'd' + var 
```

```python
## Probando para ver si la funcion logra su cometido
lista = ['ca','ce', 'ci', 'co','cu']
for item in lista:
    print(funcion(item))
```

```python
lista = datos_disney.loc[:,'show_id'] ## Copiar la columna show_id para poder modificarla esto con el fin de evitar errores en los datos
##Esto nos devolvera una serie y la transformamos a lista con el metodo list()
lista = list(lista)
```

```python
##creamos una lista vacia para agregar los valores que modifiquemos
lista_transformada = [] ##Ya tenemos la lista ahora aplicamos la función
for item in lista:
    lista_transformada = lista_transformada + [funcion(item)]
```

```python
##Eliminamos la columna show_id
del datos_disney['show_id']
```

```python
## Acá agregaré la lista ya transformada 
datos_disney['Id'] = pd.Series(lista_transformada)
```

```python
#visualizamos
datos_disney
```

Los valores nulos del campo rating deberán reemplazarse por el string “G” (corresponde al maturity rating: “general for all audiences”

```python
##Para reemplazar los valores null en la columna rating (reemplazamos por la letra G)
datos_disney['rating'].fillna('G', inplace=True)
```

```python
##Comprobamos para ver si existen valores nulos en la columna rating
datos_disney['rating'].isna().sum()
```

De haber fechas, deberán tener el formato AAAA-mm-dd

```python
## se extraera la columna date_added al formato AAAA-MM-DD

fechas = pd.to_datetime(datos_disney['date_added'])
fechas = list(fechas)
del datos_disney['date_added']
datos_disney['date_added'] = pd.Series(fechas)
```

```python
datos_disney.head(12)  
```

Los campos de texto deberán estar en minúsculas, sin excepciones

```python
##se usara el metodo lower para cambiar todos los textos a minusculas
datos_disney['type'] = datos_disney['type'].str.lower()
datos_disney['title'] = datos_disney['title'].str.lower()
datos_disney['director'] = datos_disney['director'].str.lower()
datos_disney['cast'] = datos_disney['cast'].str.lower()
datos_disney['country'] = datos_disney['country'].str.lower()
datos_disney['listed_in'] = datos_disney['listed_in'].str.lower()
datos_disney['description'] = datos_disney['description'].str.lower()
```

El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

```python
##Uso el metodo split para dividir la columna duration 
datos_disney[['duration_int','duration_type']] = datos_disney['duration'].str.split(expand=True)
```

```python
##chusmeamos :P
datos_disney.head(10)
```
